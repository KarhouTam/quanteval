# ⚔️ Strategy Comparison / 策略对比

Which strategy is the winner today? 🏁

In this notebook, we'll put different strategies head-to-head on the same dataset and even see if combining them makes for a smoother ride! 🧘

哪个策略才是今天的赢家？🏁

在本 Notebook 中，我们将让不同的策略在相同的数据集上一决高下，甚至看看把它们组合在一起是否能让收益曲线更平稳！🧘


```python
import numpy as np
import pandas as pd

from quanteval import BollingerMeanReversionStrategy, DualMAStrategy, StrategyComparator
```

```python
index = pd.date_range('2022-01-03', periods=220, freq='B')
close = pd.Series(120 + np.linspace(0, 20, len(index)) + np.sin(np.arange(len(index)) / 4) * 4, index=index)
data = pd.DataFrame({
    'Open': close * 0.99,
    'High': close * 1.01,
    'Low': close * 0.98,
    'Close': close,
    'Volume': np.linspace(1500, 2400, len(index)),
    'Amount': close * 1200,
}, index=index)
data['Ret'] = data['Close'].pct_change().fillna(0.0)
```

### ⚔️ 1. Battle of the Strategies / 策略一决高下

Let's compare **Dual Moving Average** with **Bollinger Mean Reversion**. 🤖

我们将让**双均线策略**和**布林带均值回归策略**来一场正面对抗。🤖


```python
comparator = StrategyComparator(
    strategies=[DualMAStrategy(5, 20), BollingerMeanReversionStrategy(20, 2.0)],
    strategy_names=['DualMA(5,20)', 'BollingerMR'],
    verbose=False,
)
comparison = comparator.compare(data)
comparison.metrics_df
```

### 🧩 2. Building a Portfolio / 构建投资组合

Why pick one winner when you can have both? 🤝 
Equally weighting them can sometimes lower your risk (and your stress levels).

为什么非要二选一呢？🤝 
通过等权组合，我们有时可以降低整体风险，让持仓过程更轻松。🧘


```python
portfolio = comparator.create_equal_weight_portfolio(data)
portfolio.summary()
```
